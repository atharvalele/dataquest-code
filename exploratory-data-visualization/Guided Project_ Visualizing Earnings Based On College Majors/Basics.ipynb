```python
%matplotlib inline
import pandas as pd
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads.describe()
```

```python
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print(raw_data_count, cleaned_data_count)
```

```python
# Sample size and median
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
```

```python
# Sample size and unemployment
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
```

```python
# Full_time and median
recent_grads.plot(x='Full_time', y='Median', kind='scatter')
```

```python
# ShareWomen and Unemployment_rate
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
```

```python
# Men and median
recent_grads.plot(x='Men', y='Median', kind='scatter')
```

```python
# Women and median
recent_grads.plot(x='Women', y='Median', kind='scatter')
```

```python
# Major and median
recent_grads.plot(x='Major_code', y='Median', kind='scatter')
```

- From the median vs major_code plot, we can see that popular majors make about 10,000 more than the average with some even almost doubling
- Women make slightly less money than Men do

```python
import matplotlib.pyplot as plt
columns = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 
           'Unemployment_rate', 'Men', 'Women']
fig = plt.figure(figsize = (10, 25))
for i in range(1, 9):
    ax = fig.add_subplot(4, 2, i)
    ax = recent_grads[columns[i-1]].hist(bins=20, xrot=90)
    ax.set_title(columns[i-1])
```

- Median salary range is about $35,000
